```python

import json
empty, non_empty = 0, 0
with open('data/diverse_reasoning_results.jsonl') as f:
    for line in f:
        data = json.loads(line)
        if len(data.get('selected_samples', [])) == 0:
            empty += 1
        else:
            non_empty += 1
print(f'Số dòng bị rỗng (F1 < 85): {empty}')
print(f'Số dòng CÓ kết quả (F1 >= 85): {non_empty}')
```
