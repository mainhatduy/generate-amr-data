```python
import json
import sys
import os
sys.path.append(os.path.abspath(".."))
from services.amr_hint.prompt_builder import build_prompt
with open("../data/systhetic_data.jsonl", "r") as f:
    data = [json.loads(line) for line in f if line.strip()]
```

```python
data[0]
```

```python
import re

def clean_model_respose(response):
    
    response = re.sub(r"<amr>.*?</amr>", "", response, flags=re.DOTALL)
    response = response.replace("<think>", "").replace("</think>", "")
    return response.strip()
```

```python
new_data = []
for i in data:
    reasoning = clean_model_respose(i.get("model_respose", [""])[0])
    
    new_data.append({
        "amr": i.get("amr", ""),
        "sentence": i.get("sentence", ""),
        "system_prompt": build_prompt(i.get("sentence", ""), i.get("amr", "")),
        "reasoning": reasoning
    })
```

```python
new_data[0]
```

```python
import pandas as pd
new_data = pd.DataFrame(new_data)
new_data.to_json("../data/processed_data.jsonl", orient="records", lines=True)
```
